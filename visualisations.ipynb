```python
### Creating a line-plot showing the development of the stocks over time grouped by industry-sector
### Preparing the data

industry_hist_cleaned = industry_hist.sort_values(['Industry', 'Date']).reset_index(drop=True).copy()

# Missing stock-data for single days result in outlier-values after grouping by industry-sector (values, which are too low and not representative)
# Filtering out those values (as far as possible) and replacing them with a more representative value (= value of the previous day)
for industry in industry_hist_cleaned['Industry'].unique(): # looping over each industry-sector
    industry_mask = industry_hist_cleaned['Industry'] == industry
    industry_data = industry_hist_cleaned[industry_mask]
    pct_change = industry_data['Close_Price'].pct_change(periods=2)
    # calculating the change-perantage in comparison to the value in the row before the previous one (period=2),
    # because a comparison with (only) the previous day still results in outliers.
    
    outlier_indices = industry_data.index[abs(pct_change) > 30/100] # defining change over 30% as an outlier and getting the indices of the outliers for each industry
    
    for i in outlier_indices:
        prev_day_index = i - 1
        
        # repalacing the outliers with the values of the previous day
        if prev_day_index >= 0 and industry_hist_cleaned.loc[prev_day_index, 'Industry'] == industry:
            prev_close_price = industry_hist_cleaned.loc[prev_day_index, 'Close_Price']
            industry_hist_cleaned.loc[i, 'Close_Price'] = prev_close_price
    
industry_hist_cleaned = industry_hist_cleaned.sort_values(['Date']).reset_index(drop=True)

# pivoting the table, to show the values for each day with one column for each industry-sector
industry_hist_cleaned_pivot = industry_hist_cleaned.pivot(
        index='Date',
        columns='Industry',
        values='Close_Price'
   )

# The missing stock data results in some NaN-values and some entries which couldn't be assigned to a industry-sector ("Not available"), which are filtered out
industry_hist_cleaned_pivot = industry_hist_cleaned_pivot.drop(columns=['Not available'])
industry_hist_cleaned_pivot = industry_hist_cleaned_pivot.dropna()

display(industry_hist_cleaned_pivot)
```

```python
### Plotting
# Due to the smoothing-out of the outliers and filtering out the NaN-values, the plot is not completely precise, but represents the trend for each industry-sector
industry_hist_cleaned_pivot.plot(figsize=(15, 10))
plt.title('Industry Development')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Closing Price (sum)', fontsize=14)
plt.grid(axis='y')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
### Reading the csv-file produced by a SQL-Query into a Data-Frame and displaying it
euro_hist = pd.read_csv('eurostoxx_historic.csv', delimiter=',')

# Converting the Data-column into a timestamp-format and setting it as the new index
euro_hist['Date'] = pd.to_datetime(euro_hist['Date'])
euro_hist.set_index('Date', inplace=True)
display(euro_hist)
```

```python
## Creating a data-frame showing the combined stock-price for each industry-sector at the end of each year (closing price)
# creating a table with yearly data

# .resample('YE') creates yearly bins and .last() selects the last value of each bin
# the industry_hist_yearly dataframe is based upon the cleaned table, therefore the data is not entirely precise
industry_hist_yearly = industry_hist_cleaned_pivot.resample('YE').last()
industry_hist_yearly['Sum All Industries'] = industry_hist_yearly.sum(axis=1)

euro_hist_yearly = euro_hist.resample('YE').last()

# merging the tables
yearly_all = industry_hist_yearly.join(euro_hist_yearly, on='Date')
yearly_all = yearly_all.rename(columns={'Close Price End': 'EURO STOXX 50'})

display(yearly_all)
```

```python
# displaying the change to each previous Year in %
yearly_all_pct = yearly_all.pct_change()
yearly_all_pct = yearly_all_pct.dropna()
yearly_all_pct.style.format("{:.2%}")
```

```python
## Creating bar-subplots for each year displaying the change of the combined stock price for each sector in comparison to the previous year

fig, axes = plt.subplots(2, 2, figsize=(22, 18), sharey=True)
colors = ['blue', 'beige', 'red', 'purple', 'navy', 'brown', 'pink', 'yellow', 'cyan', 'magenta', 'orange', 'green']
year_index = 0
for i in range(2):
    for n in range(2):
        ax = axes[i, n]
        ax.bar(yearly_all_pct.columns, yearly_all_pct.iloc[year_index], color=colors)
        ax.set_title(str(yearly_all_pct.index[year_index].year), fontsize=14)
        ax.set_ylabel('Percentage Change to previous year', fontsize=14)
        ax.tick_params(axis='x', rotation=45)
        ax.grid(axis='y', alpha=0.9)
        year_index += 1
plt.tight_layout()
plt.show()
```
